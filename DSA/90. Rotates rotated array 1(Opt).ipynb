```python
l = [17,18,20,1,3,4,5,7,8,10,11,13,14,16]
target = 4
def find(l,target): 
    n = len(l)
    low,high = 0, n-1
    while low <= high: 
        mid = (low + high) // 2
        if l[mid] == target: 
            return mid
        if l[mid] <= l[high]: 
            if l[mid] <= target <= l[high]: 
                low = mid + 1
            else: 
                high = mid - 1
        else: 
            if l[low] <= target <= l[high]: 
                high = mid - 1
            else: 
                low = mid + 1
        return -1

print(find(l,target))
```
