```python
def lower_bound(L,target):
    n = len(L)
    lb = n
    low, high = 0, n-1
    while low < high: 
        mid = (low + high) // 2
        if L[mid] >= target: 
            lb = mid
            high = mid - 1
        else: 
            low = mid + 1
    return lb

def upper_bound(L,target): 
    n = len(L)
    ub = n
    low,high = 0, n-1
    while low <= high: 
        mid = (low + high) // 2
        if L[mid] > target: 
            ub = mid
            high = mid - 1
        else: 
            low = mid + 1
    return ub

L = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3

lb = lower_bound(L,target)
if lb == -1: 
    print([-1,-1])
ub = upper_bound(L,target)

print([lb,ub-1])
```
