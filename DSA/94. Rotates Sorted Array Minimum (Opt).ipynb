```python
def find_minimum(nums):
    n = len(nums)
    low, high = 0, n-1
    mini = float('inf')
    while low < high:
        mid = (low + high) // 2
        if nums[mid] <= nums[high]:
            mini = min(mini, nums[mid])
            high = mid - 1
        else: 
            mini = min(mini, nums[low])
            low = mid + 1
    return mini

nums = [7,8,9,1,2,3,4]
print(find_minimum(nums))
```
