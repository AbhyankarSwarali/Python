```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None

class ReverseLinkedList: 
    def __init__(self): 
        self.head = None

    def reverse_list(self): 
        temp = self.head
        stack = []
        
        while temp is not None: 
            stack.append(temp.data)
            temp = temp.next
        temp = self.head
        
        while temp is not None: 
            e = stack.pop()
            temp.data = e
            temp = temp.next
            
        return self.head.data

node1 = Node(5)
node2 = Node(8)
node3 = Node(7)
node4 = Node(3)
node5 = Node(2)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

reverse = ReverseLinkedList()
reverse.head = node1

print(reverse.reverse_list())
```
