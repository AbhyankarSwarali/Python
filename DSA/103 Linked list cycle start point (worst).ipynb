```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None

class LinkedListCycle: 
    def __init__(self): 
        self.head = None

    def cycle_start_point(self): 
        temp = self.head
        my_set = set()
        while temp is not None: 
            if temp in my_set: 
                return temp
            my_set.add(temp)
            temp = temp.next
        return 'No cycle present'

node1 = Node(5)
node2 = Node(9)
node3 = Node(1)
node4 = Node(7)
node5 = Node(6)
node6 = Node(4)
node7 = Node(9)
node8 = Node(2)
node9 = Node(8)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8
node8.next = node9
node9.next = node3

cycle = LinkedListCycle()
cycle.head = node1

print(cycle.cycle_start_point())
```
