```python
def lower_bound(L,target):
    n = len(L)
    lb = n
    low, high = 0, n-1
    while low < high: 
        mid = (low + high) // 2
        if L[mid] >= target: 
            lb = mid
            high = mid - 1
        else: 
            low = mid + 1
    return lb

def upper_bound(L,target): 
    n = len(L)
    ub = n
    low,high = 0, n-1
    while low <= high: 
        mid = (low + high) // 2
        if L[mid] > target: 
            ub = mid
            high = mid - 1
        else: 
            low = mid + 1
    return ub

l = [1,2,3,3,3,3,3,5,6,8,9,9,10,12,12,12]
target = 12
lb = lower_bound(l,target)
if lb == -1: 
    print(0)
ub = upper_bound(l,target)

print(ub-lb)
```
