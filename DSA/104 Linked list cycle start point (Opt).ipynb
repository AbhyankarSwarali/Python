```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None

class LinkedListCycle: 
    def __init__(self): 
        self.head = None

    def cycle_start_point(self): 
        slow = self.head
        fast = self.head
        while fast and fast.next is not None: 
            slow = slow.next
            fast = fast.next.next
            if slow == fast: 
                slow = self.head
                while slow != fast: 
                    slow = slow.next
                    fast = fast.next
                return slow
        return None

node1 = Node(5)
node2 = Node(9)
node3 = Node(1)
node4 = Node(7)
node5 = Node(6)
node6 = Node(4)
node7 = Node(9)
node8 = Node(2)
node9 = Node(8)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8
node8.next = node9
node9.next = node3

cycle = LinkedListCycle()
cycle.head = node1

print(cycle.cycle_start_point())
```
