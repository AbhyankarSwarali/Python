```python
class Node: 
    def __init__(self,data): 
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList: 
    def __init__(self): 
        self.head = None

    def pairs_sum(self):
        target = 7
        result = []
        left = self.head
        right = self.head
        while right.next is not None: 
            right = right.next
        while left is not None and right is not None and left.data < right.data: 
            total = left.data + right.data
            if total == target: 
                result.append([left.data,right.data])
                left = left.next
                right = right.prev
            elif total > target: 
                right = right.prev
            else: 
                left = left.next

        return result
        
n1 = Node(1)
n2 = Node(2)
n3 = Node(4)
n4 = Node(5)
n5 = Node(6)
n6 = Node(8)
n7 = Node(9)

n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
n7.prev = n6
n6.prev = n5
n5.prev = n4
n4.prev = n3
n3.prev = n2
n2.prev = n1

dll = DoublyLinkedList()
dll.head = n1
print(dll.pairs_sum())
```
