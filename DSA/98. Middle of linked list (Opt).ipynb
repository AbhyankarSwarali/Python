```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None

class LinkedList: 
    def __init__(self): 
        self.head = None

    def find_middle(self): 
        slow = self.head
        fast = self.head

        while fast is not None and fast.next is not None: 
            slow = slow.next
            fast = fast.next.next

        return slow.data

node1 = Node(5)
node2 = Node(10)
node3 = Node(27)
node4 = Node(3)
node5 = Node(99)
node6 = Node(13)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6

L1 = LinkedList()
L1.head = node1

print(L1.find_middle())
```
