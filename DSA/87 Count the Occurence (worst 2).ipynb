```python
def occurence_count(l,target):
    n = len(l)
    first = -1
    last = -1
    for i in range(n): 
        if l[i] == target: 
            if first == -1: 
                first = i
            last = i
    if first == -1: 
        return 0
    return (last - first + 1)

l = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
print(occurence_count(l,target))
```
