```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None
        self.prev = None

class DoublylinedList: 
    def __init__(self): 
        self.head = None

    def delete_k(self): 
        k = 2
        if self.head.next is None and self.head.data == k: 
            return None

        temp = self.head
        prev = None
        new_head = self.head
        while temp is not None: 
            if temp.data == k: 
                if prev is not None: 
                    prev.next = temp.next
                if temp.next is not None:
                    temp.next.prev = prev
                if temp == new_head: 
                    new_head = new_head.next
            prev = temp
            temp = temp.next
        return new_head.data

n1 = Node(2)
n2 = Node(2)
n3 = Node(3)
n4 = Node(2)
n5 = Node(10)

n1.next = n2
n2.next = n3
n3.n3xt = n4
n4.next = n5
n5.prev = n4
n4.prev = n3
n3.prev = n2
n2.prev = n1

dll = DoublylinedList()
dll.head = n1
print(dll.delete_k())
```
