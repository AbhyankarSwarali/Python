```python
def check_elements(l, target): 
    low, high = 0, len(l)-1
    while low <= high: 
        mid = (low + high) // 2
        if l[mid] == target: 
            return True
        if l[low] == l[mid] == l[high]: 
            low += 1
            high -= 1
            continue
        if l[low] <= l[mid]:
            if l[low] <= target <= l[mid]: 
                high = mid - 1
            else: 
                low = mid + 1
        else: 
            if l[mid] <= target <= l[high]: 
                low = mid + 1
            else: 
                high = mid - 1
                
    return False

l = [7,7,7,7,7,7,7,7,1,2,3,4,5,7,7]
target = 3
print(check_elements(l,target))
```
