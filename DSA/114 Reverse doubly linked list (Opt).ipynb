```python
class Node: 
    def __init__(self, data): 
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList: 
    def __init__(self): 
        self.head = None

    def reverse(self): 
        if self.head is None: 
            return self.head.data
            
        current = self.head
        previous = None
        
        while current is not None: 
            front = current.next
            current.next = previous
            current.prev = front
            previous = current
            current = front
            
        return previous.data

n1 = Node(5)
n2 = Node(3)
n3 = Node(2)
n4 = Node(1)
n5 = Node(9)

n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.prev = n4
n4.prev = n3
n3.prev = n2
n2.prev = n1

dll = DoublyLinkedList()
dll.head = n1
print(dll.reverse())
```
