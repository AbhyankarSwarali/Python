```python
class Node:
    def __init__(self,data): 
        self.data = data
        self.next = None

node1 = Node(5)
node2 = Node(10)
node3 = Node(7)
node4 = Node(8)

node1.next = node2
node2.next = node3
node3.next = node4

print(node1.data)
print(node1.next.data)
print(node1.next.next.next.data)

print(node1.next)
print(node2)
```
